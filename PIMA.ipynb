```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
x=df.iloc[:,0:8]
y=df.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python

from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
df_label = df['Outcome']
df_features = df.drop('Outcome', 1)
df_features.replace('?', -99999, inplace=True)
```

```python
label = []
for lab in df_label:
    if lab == 1:
        label.append([1, 0])  # class 1
    elif lab == 0:
        label.append([0, 1])  # class 0
```

```python
feature_names = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
data = df[feature_names]
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, label, test_size=0.2, random_state=42)
```

```python
model = Sequential()
model.add(Dense(500, input_dim=8, activation='sigmoid'))
model.add(Dense(100, activation='sigmoid'))
model.add(Dense(2, activation='softmax'))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(np.array(x_train),np.array(y_train), epochs=1000, batch_size=70, validation_data=(np.array(x_test),np.array(y_test)))
```
